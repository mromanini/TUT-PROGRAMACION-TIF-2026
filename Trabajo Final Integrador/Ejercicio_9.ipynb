```python
import networkx as nx
grafod = nx.DiGraph()
grafod.add_nodes_from(["A","B","C","D","E","F","G","H"])
grafod.add_edges_from([("A","B"),("A","C"),("B","C"),("B","D"),("C","D"),("C","E"),("D","E"),("D","F"),("E","F"),("E","G"),("F","G"),("F","H"),("G","A"),("G","H"),("H","A"),("H","B")])
nx.draw(grafod,with_labels=True,font_weight='bold', pos=nx.spectral_layout(grafod),node_color='skyblue',node_size=800)
```
